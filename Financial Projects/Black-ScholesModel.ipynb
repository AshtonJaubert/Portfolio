```python
import math
from scipy.stats import norm
```

```python
S = 45 #Underlying Price
K = 40 # Strike Price
T = 2 #Time to Expiration
r = .1 # Risk-Free Rate
vol = 0.5 # Volatility 
```

# Identify D1 and D2 variables:

```python
d1 = (math.log(S/K) + (r + 0.5 * vol**2)*T ) / (vol * math.sqrt(T))
```

```python
d2 = d1 - (vol * math.sqrt(T))
```

Now we can Identify The Call and Put Option Price

## Calculate Call Option Price


```python
C = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
```

## Calculate Put Option Price:

```python
P = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
```

# The Results:

```python
print('value of d1 is:', round(d1,4))
print('value of d2 is:', round(d2,4))
print(' Price of the Call Option', round(C,2))
print(' Price of the Put Option', round(P,2))
```
